```python
def remove_duplicate(num:int)->int:
    if not isinstance(num,list):
         raise TypeError("Must be a list.")
    
    for item in num:
        if not isinstance(item,(int,float)):
            raise ValueError(" Must be a numeri .")

            
    list2 = []
    for item in num:
        if not item in list2:
            list2.append(item)
    return list2 


if __name__ == "__main__":
    try:
        data = input("Enter the list:")  
        data = data.strip("[]")
        if data == "":
            print("Empty")
        # elif len(data) <= 2 :
        #     print("list must be greater than 2 element.")
        
        else:
            user_input = list(map(int,data.split(",")))
            if len(user_input) <= 2:
                print("must be two element in list.")
            else:
                print(remove_duplicate(user_input))
            
            
            
    except TypeError as e:
        print("only for list",e)
    except IndexError as e:
        print("list item length is greater than 2.",e)
    except ValueError as e:
        print("only for numeric.",e)
    except Exception as e:
        print(e)
        
```
